```python
import pandas as pd
import os
```

```python
def usa_f_open (path,sep):
    data3 = open (path,"r")
    cols = data3.readline().strip().split(sep)
    n_cols = len (cols)
    counter = 0
    main_dict ={}
    for col in cols:
        main_dict[col]=[]
        
    for line in data3:
        values = line.strip().split(sep)
        for i in range (len(cols)):
            main_dict [cols[i]].append(values[i])
        counter +=1
        
    print("El data set tiene %d filas y %d columnas"%(counter, n_cols))
    df = pd.DataFrame(main_dict)
    return df
```

```python
df_prueba= usa_f_open("../datasets/customer-churn-model/Customer Churn Model.csv",',')
df_prueba.head(3)
```
